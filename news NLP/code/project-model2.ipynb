```python
import re
from string import punctuation
#Making all text lowercase, removing punctuation and extra spaces
def normalization(text):
    #Converting descriptions to lowercase
    text = str(text)
    Lowercase = text.lower()
    clean_text = r''
    new = punctuation + clean_text
    #Removeing punctuation by defining a pattern to match punctuation
    Nopunctuation = re.compile('[{}]'.format(re.escape(new)))

    #Removing extra spaces by sliting text and joining with spaces to remove extra spaces
    return " ".join(Nopunctuation.split(Lowercase))


column_to_clean = 'headline'
dataset[column_to_clean] = dataset[column_to_clean].apply(normalization)
dataset
```

```python
#Re-indexing the rows
new_index = range(len(dataset))  # For example, reindexing with a range of numbers
dataset = dataset.reindex(new_index)
#Saving normalized dataset
output_file_name = 'normalized.xlsx'
dataset.to_excel(output_file_name, index=False)
```

step 2: stop word removal

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')

def remove_stopwords(text):
    tokens = nltk.word_tokenize(text)
    required_words = [x for x in tokens if x not in set(stopwords.words('english'))]

    sentence_with_no_stopwords = ' '.join(required_words)

    return sentence_with_no_stopwords


dataset['headline'] = dataset['headline'].astype(str)

#dataset['short_description'] = dataset['short_description'].apply(lambda x: remove_stopwords(x))
dataset['headline'] = dataset['headline'].apply(remove_stopwords)

dataset

output_file_name = 'removed_stopwords.xlsx'
dataset.to_excel(output_file_name, index=False)
```

step 3: lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

def lemmitization(text):
    token = text.split()
    required_world = [WordNetLemmatizer().lemmatize(x) for x in token]
    sentence = " ".join(required_world)

    return sentence

dataset['headline'] = dataset['headline'].apply(lemmitization)

output_file_name = 'lemmatized.xlsx'
dataset.to_excel(output_file_name, index=False)
dataset
```

step 4: stemming

```python
from nltk.stem import PorterStemmer

def stemming(text):
    token = word_tokenize(text)

    required_words = [PorterStemmer().stem(x) for x in token]
    sentence = ' '.join(required_words)

    return sentence

dataset.loc[:,'headline'] = dataset['headline'].apply(stemming)
output_file_name = 'stemmed.xlsx'
dataset.to_excel(output_file_name, index=False)
```

```python
category_counts = dataset['category'].value_counts()

#Filter out categories with less than 3500 rows
categories_to_keep = category_counts[category_counts >= 3500].index

#Keep only the rows where the category is in categories_to_keep
dataset = dataset[dataset['category'].isin(categories_to_keep)]


#Limit each category to 5000 rows
def limit_rows(group):
    return group.head(3500)

#Apply function to each categories
dataset = dataset.groupby('category', group_keys=False).apply(limit_rows)
```

step 5: vectorization

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dataset['headline'], dataset['category'], test_size = 0.2, random_state = 53)
```

```python
from sklearn.feature_extraction.text import HashingVectorizer

#Fitting the vectorizer on training data
vectorizer = HashingVectorizer(n_features=9100, alternate_sign=False)
X_train = vectorizer.fit_transform(X_train)

#Transform the testing data using the same vectorizer
X_test = vectorizer.transform(X_test)
```

step 6: text classification

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Create a KNN classifier
knn = KNeighborsClassifier(n_neighbors=119)

# Train the classifier
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, classification_report

#X_test = vectorizer.transform(X_test)
y_pred = knn.predict(X_test)

print(confusion_matrix(y_test, y_pred))

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```
