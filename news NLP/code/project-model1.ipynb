```python
import re
from string import punctuation
#Making all text lowercase, removing punctuation and extra spaces
def normalization(text):
    #Converting descriptions to lowercase
    text = str(text)
    Lowercase = text.lower()
    clean_text = r''
    new = punctuation + clean_text
    #Removeing punctuation by defining a pattern to match punctuation
    Nopunctuation = re.compile('[{}]'.format(re.escape(new)))

    #Removing extra spaces by sliting text and joining with spaces to remove extra spaces
    return " ".join(Nopunctuation.split(Lowercase))


column_to_clean = 'short_description'
dataset[column_to_clean] = dataset[column_to_clean].apply(normalization)
dataset
```

```python
#Re-indexing the rows
new_index = range(len(dataset))  # For example, reindexing with a range of numbers
dataset = dataset.reindex(new_index)
#Saving normalized dataset
output_file_name = 'normalized.xlsx'
dataset.to_excel(output_file_name, index=False)
```

step 2: stop word removal

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')

def remove_stopwords(text):
    tokens = nltk.word_tokenize(text)
    required_words = [x for x in tokens if x not in set(stopwords.words('english'))]

    sentence_with_no_stopwords = ' '.join(required_words)

    return sentence_with_no_stopwords

dataset['short_description'] = dataset['short_description'].astype(str)
dataset['short_description'] = dataset['short_description'].apply(remove_stopwords)

dataset

output_file_name = 'removed_stopwords.xlsx'
dataset.to_excel(output_file_name, index=False)
```

step 3: lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

def lemmitization(text):
    token = text.split()
    required_world = [WordNetLemmatizer().lemmatize(x) for x in token]
    sentence = " ".join(required_world)

    return sentence

dataset['short_description'] = dataset['short_description'].apply(lemmitization)

output_file_name = 'lemmatized.xlsx'
dataset.to_excel(output_file_name, index=False)
dataset
```

step 4: stemming

```python
from nltk.stem import PorterStemmer

def stemming(text):
    token = word_tokenize(text)

    required_words = [PorterStemmer().stem(x) for x in token]
    sentence = ' '.join(required_words)

    return sentence

dataset.loc[:,'short_description'] = dataset['short_description'].apply(stemming)
output_file_name = 'stemmed.xlsx'
dataset.to_excel(output_file_name, index=False)
```

step 5: vectorization

```python
from gensim.models import Word2Vec
import numpy as np
np.random.seed(26)

#Importing and using pre-trained word vectors
import gensim.downloader as api
print(list(api.info()['models'].keys()))

wv = api.load('word2vec-google-news-300')

dataset['text'] = dataset['short_description'].apply(word_tokenize)

def vectorizer(text):
    vectorsize = wv.vector_size
    wv1 = np.zeros(vectorsize)
    ctr = 1
    for word in text:
        if word in wv:
            ctr += 1
            wv1 += wv[word]
    wv1 = wv1/ctr
    return wv1

dataset['vector'] = dataset['text'].apply(vectorizer)
```

step 6: text classification

```python
X = dataset['vector'].to_list()
y = dataset['category'].to_list()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
X = [str(text) for text in X]

#Tokenizing text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
word_index = tokenizer.word_index
vocab_size = len(word_index) + 1

#Converting text data to sequences
sequences = tokenizer.texts_to_sequences(X)
max_sequence_length = max([len(seq) for seq in sequences])
padded_sequences = pad_sequences(sequences, maxlen=max_sequence_length, padding='post')

#Converting labels to one-hot encoding
y_hot = tf.keras.utils.to_categorical(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, y_hot, test_size=0.2, random_state=42)
```

```python
from keras import layers
from tensorflow.keras.layers import LSTM, Dense, Embedding, Bidirectional

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 100, input_length=max_sequence_length),
    tf.keras.layers.Bidirectional(LSTM(units=100, dropout=0.2, recurrent_dropout=0.2)),
    tf.keras.layers.Dense(487, activation='relu'),
    tf.keras.layers.Dense(150, activation='relu'),
    tf.keras.layers.Dense(28, activation='relu'),
    tf.keras.layers.Dense(12, activation='softmax')  
])
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

import numpy as np

X_train = np.array(X_train)
y_train = np.array(y_train)

X_test = np.array(X_test)
y_test = np.array(y_test)

# Train the model
model.fit(X_train, y_train, epochs=6, batch_size=128, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report


# Predict classes for test data
y_pred = np.argmax(model.predict(X_test), axis=1)

# Generate confusion matrix
conf_matrix = confusion_matrix(np.argmax(y_test, axis=1), y_pred)

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=range(12), yticklabels=range(12))
plt.xlabel('Predicted category')
plt.ylabel('True category')
plt.title('Confusion Matrix')
plt.show()

# Print classification report
print(classification_report(np.argmax(y_test, axis=1), y_pred))
```
